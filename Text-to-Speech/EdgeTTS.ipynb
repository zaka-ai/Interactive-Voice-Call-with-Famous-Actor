## 2. EdgeTTS Class

```python
import edge_tts
import asyncio
import io
from IPython.display import Audio, display
import nest_asyncio

class EdgeTTS:
    """Simple class to handle Edge TTS functionality."""

    async def generate_speech(self, text, voice="ar-OM-AyshaNeural"):
        """Generate speech and return audio data"""
        communicate = edge_tts.Communicate(text, voice)

        audio_data = io.BytesIO()
        async for chunk in communicate.stream():
            if chunk["type"] == "audio":
                audio_data.write(chunk["data"])

        audio_data.seek(0)
        return audio_data

    def speak_text_colab(self, text, voice="ar-OM-AyshaNeural"):
        """Speak text in Colab using IPython Audio display"""
        # Get the current event loop
        loop = asyncio.get_event_loop()
        audio_data = loop.run_until_complete(self.generate_speech(text, voice))

        # Display audio player in Colab
        display(Audio(audio_data.read(), autoplay=True))

        # Return to beginning of stream if you want to use it again
        audio_data.seek(0)
        return audio_data
```
